#Data Extraction

```python
mask_img = os.listdir(mask_dir)
nomask_img = os.listdir(nomask_dir)
```

```python
data_mask = []
for i in mask_img:
    data_mask.append(mask_dir+i)
```

```python
data_nonmask = []
for i in nomask_img:
    data_nonmask.append(nomask_dir+i)
```

```python
dataset = []
labels = []
for i in data_mask:
  print(i)
  label = 1
  ime = cv2.imread(i)
  if(ime is not None):
    ime = cv2.cvtColor(ime, cv2.COLOR_BGR2GRAY)
    ime = cv2.resize(ime, (64,64))
    ime = ime.flatten()
    dataset.append(ime)
    labels.append(label)
```

```python
dataset1 = []
labels1 = []
for i in data_nonmask:
  print(i)
  label = 0
  ime = cv2.imread(i)
  if(ime is not None):
    ime = cv2.cvtColor(ime, cv2.COLOR_BGR2GRAY)
    ime = cv2.resize(ime, (64,64))
    ime = ime.flatten()
    dataset1.append(ime)
    labels1.append(label)
```
